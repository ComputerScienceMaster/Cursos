## Criando nosso primeiro gráfico de barras

Agora vamos criar nosso primeiro gráfico de barras simples utilizando algumas opções básicas do matplotlib. Mas antes, vamos importar a biblioteca usando o comando de importação.

```python
import matplotlib.pyplot as plt
```

A seguir, vamos criar um gráfico de barras simples usando a idade das pessoas como parâmetro.

```python
plt.bar(tbl['Cidade'], tbl['Idade'])
plt.xticks(tbl['Cidade'], tbl['Nome'])
plt.show()
```

Agora que nosso primeiro gráfico de linhas está pronto, podemos perceber que existem alguns problemas. 
O primeiro deles é que nosso gráfico parece muito pequeno (isso prejudica a visualização dos labels)...

Vamos resolver isso com o comando figsize

```python
plt.figure(figsize=(10,6))
plt.bar(tbl['Cidade'], tbl['Idade'])
plt.xticks(tbl['Cidade'], tbl['Nome'])
plt.show()
```

Infelizmente, apenas ajustar o tamanho do gráfico não parece resolver o problema. Vamos tentar rotacionar os labels em 90 graus.

```python
plt.figure(figsize=(10,6))
plt.bar(tbl['Cidade'], tbl['Idade'])
plt.xticks(tbl['Cidade'], tbl['Nome'], rotation='vertical')
plt.show()
```

Beeeem melhor não é mesmo? agora podemos adicionar alguns elementos a mais, como por exemplo o título e algumas linhas de grade:

```python
plt.figure(figsize=(10,6))
# adicionando o título
plt.title("Pesquisa de Opinião")
## adicionando alguns grids
plt.grid(True, color = "grey", linewidth = "1", linestyle = "-.")
plt.bar(tbl['Cidade'], tbl['Idade'])
plt.xticks(tbl['Cidade'], tbl['Nome'], rotation='vertical')
plt.show()
```

Vamos agora alterar as cores das barras do nosso gráfico:

```python
plt.figure(figsize=(10,6))
# adicionando o título
plt.title("Pesquisa de Opinião")
## adicionando alguns grids
plt.grid(True, color = "grey", linewidth = "1", linestyle = "-.")
plt.bar(tbl['Cidade'], tbl['Idade'],  color="k")
plt.xticks(tbl['Cidade'], tbl['Nome'], rotation='vertical')
plt.show()
```

# Gráfico de barras com duas séries

Agora que já criamos nosso gráfico de barras simples, vamos criar um gráfico de barras com duas séries no mesmo gráfico. Para isso vamos usar

```python
import numpy as np

plt.figure(figsize=(10,6))
def subcategorybar(X, vals, width=.7):
    n = len(vals)
    _X = np.arange(len(X))
    for i in range(n):
        plt.bar(_X - width/2. + i/float(n)*width, vals[i], width=width/float(n), align="edge")   
    plt.xticks(_X, X, rotation="vertical")
    
subcategorybar(tbl['Nome'][0:6], [tbl['Peso'][0:6],tbl['Idade'][0:6]])
plt.grid(True, color = "grey", linewidth = "1", linestyle = "-.")
plt.show()
```

Agora vamos criar o famoso histograma, um gráfico de barras em que não existe grande complexidade, porém exige o calculo de algumas frequências. Portanto, para isso podemos usar uma ferramenta auxiliar que é o pandas.

```python
# para plotar o histograma apenas selecione a série e indique a quantidade de bins
tbl['Peso'].hist(bins=10)
```
