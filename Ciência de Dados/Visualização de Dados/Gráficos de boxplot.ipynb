# Introdução

Nesse notebook vamos aprender como criar gráficos boxplot usando o pandas e o matplotlib. Abandone definitivamente o excel e vamos mergulhar nessa ferramenta de visualização incrível.

Para esse tutorial irei utilizar o dataset "pesquisa_populaçao" e vamos importar de dentro da pasta "data" o xls chamado "pesquisa_populacao". A seguir vamos chamar o comando tbl.head() para mostrar as primeiras 5 linhas do nosso dataset.

```python
import pandas as pd

tbl = pd.read_excel("data/pesquisa_populacao.xlsx")
tbl.head()
```

## Criando nosso primeiro gráfico de área

Agora vamos criar nosso primeiro gráfico boxplot (bem simples) utilizando algumas opções básicas do matplotlib. Mas antes, vamos importar a biblioteca usando o comando de importação.

```python
import matplotlib.pyplot as plt
```

A seguir, vamos criar um gráfico usando apenas os dados do salário

```python
plt.boxplot(tbl['Salário'])
# show plot
plt.show()
```

Agora que nosso primeiro gráfico boxplot está pronto, podemos perceber que existem alguns problemas. O primeiro deles é que nosso gráfico parece muito pequeno, sendo assim, vamos aumentá-lo usando o comando figsize.

```python
fig = plt.figure(figsize =(10, 7))
# Creating plot
plt.boxplot(tbl['Salário'])
# show plot
plt.show()
```

Um outro problema que percebemos logo de cara é que não existem títulos para nada, nem para os eixos, nem para o gráfico, nem para as linhas. Certo, tudo isso faz parte da legibilidade de um gráfico, sendo assim, vamos dar uma melhorada começando pelo título do gráfico. Para isso vamos executar o seguinte comando:

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(tbl['Salário'])
plt.title("Salário das pessoas" , fontsize=15)
plt.xlabel("Pessoas" , fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Agora vamos adicionar dois boxplots no mesmo gráfico

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot([tbl['Idade'], tbl['Peso']])
plt.title("Salário das pessoas" , fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Agora vamos colocar uma corzinha em nosso boxplot

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot([tbl['Idade'], tbl['Peso']],  patch_artist=True)
plt.title("Salário das pessoas" , fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Podemos ainda adicionar uma "cinturinha" em nosso boxplot para que ele fique mais fácil de ler

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot([tbl['Idade'], tbl['Peso']], notch=True, patch_artist=True)
plt.title("Salário das pessoas" , fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Agora vamos adicionar cores diferentes para nosso boxplot

```python
fig = plt.figure(figsize =(10, 7))
box = plt.boxplot([tbl['Idade'], tbl['Peso']], notch=True, patch_artist=True)
plt.title("Salário das pessoas" , fontsize=15)
plt.ylabel("Salário", fontsize=15)
colors = ['cyan', 'lightblue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.show()
```
