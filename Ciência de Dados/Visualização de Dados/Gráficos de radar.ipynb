# Introdução

Nesse notebook vamos aprender como criar gráficos de radar usando o matplotlib. Abandone definitivamente o excel e vamos mergulhar nessa ferramenta de visualização incrível.

Para esse tutorial irei utilizar alguns dados inseridos aqui mesmo em nosso notebook, então não será necessário importar nenhum dado adicional.

```python
import pandas as pd
import numpy as np
```

```python
felipe = {"Agilidade": 50, "Inteligência": 15, "Resiliência": 15, "Socialização": 29, "Habilidades manuais": 15}
paulo = {"Agilidade": 40, "Inteligência": 35, "Resiliência": 25, "Socialização": 31, "Habilidades manuais": 85}
maria = {"Agilidade": 20, "Inteligência": 15, "Resiliência": 95, "Socialização": 34, "Habilidades manuais": 97}
tulio = {"Agilidade": 10, "Inteligência": 11, "Resiliência": 35, "Socialização": 11, "Habilidades manuais": 100}
januario = {"Agilidade": 15, "Inteligência": 83, "Resiliência": 10, "Socialização": 59, "Habilidades manuais": 80}
claudio = {"Agilidade": 85, "Inteligência": 95, "Resiliência": 13, "Socialização": 90, "Habilidades manuais": 79}
vanessa = {"Agilidade": 95, "Inteligência": 65, "Resiliência": 54, "Socialização": 97, "Habilidades manuais": 46}

skills = [felipe, paulo, maria, tulio, januario, claudio, vanessa]
df = pd.DataFrame(skills)
df
```

## Criando nosso primeiro gráfico de área

Agora vamos criar nosso primeiro gráfico de radar (bem simples) utilizando algumas opções básicas do matplotlib. Mas antes, vamos importar a biblioteca usando o comando de importação.

```python
import matplotlib.pyplot as plt
```

A seguir, vamos criar um gráfico de radar usando o felipe como parametro

```python
labels = ["Agilidade", "Inteligência","Resiliência","Socialização","Habilidades manuais"]
values = df.loc[0].tolist()
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
values += values[:1]
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values, color='red', linewidth=1)
```

Agora se quisermos "pintar" a área do nosso radar fazemos:

```python
labels = ["Agilidade", "Inteligência","Resiliência","Socialização","Habilidades manuais"]
values = df.loc[0].tolist()
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
values += values[:1]
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values, color='red', linewidth=1)
ax.fill(angles, values, color='red', alpha=0.25)
```
