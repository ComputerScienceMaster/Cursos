# Introdução

Nesse notebook vamos aprender como criar gráficos de pontos usando o pandas e o matplotlib. Abandone definitivamente o excel e vamos mergulhar nessa ferramenta de visualização incrível.

Para esse tutorial irei utilizar o dataset "pesquisa_populaçao" disponibilizado que armazena o tráfego de um website. Portanto, vamos importar de dentro da pasta "data" o csv chamado "pesquisa_populacao". A seguir vamos chamar o comando dates.head() para mostrar as primeiras 5 linhas do nosso dataset.

```python
import pandas as pd

tbl = pd.read_excel("data/pesquisa_populacao.xlsx")
tbl.head()
```

## Criando nosso primeiro gráfico de área

Agora vamos criar nosso primeiro gráfico de área (bem simples) utilizando algumas opções básicas do matplotlib. Mas antes, vamos importar a biblioteca usando o comando de importação.

```python
import matplotlib.pyplot as plt
```

A seguir, vamos criar um gráfico pontos usando apenas a idade e o salário dos participantes da pesquisa.

```python
plt.scatter(tbl["Idade"], tbl["Salário"])
```

Agora que nosso primeiro gráfico de linhas está pronto, podemos perceber que existem alguns problemas. O primeiro deles é que nosso gráfico parece muito pequeno, sendo assim, vamos aumentá-lo usando o comando figsize.

```python
plt.figure(figsize=(10,6))
plt.scatter(tbl["Idade"], tbl["Salário"])
plt.show()
```

Um outro problema que percebemos logo de cara é que não existem títulos para nada, nem para os eixos, nem para o gráfico, nem para as linhas. Certo, tudo isso faz parte da legibilidade de um gráfico, sendo assim, vamos dar uma melhorada começando pelo título do gráfico. Para isso vamos executar o seguinte comando:

```python
plt.figure(figsize=(10,6))
plt.scatter(tbl["Idade"], tbl["Salário"])
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Agora vamos mudar as cores dos nossos pontinhos...

```python
plt.figure(figsize=(10,6))
plt.scatter(tbl["Idade"], tbl["Salário"], color="r")
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

E se nós quisermos seccionar esses pontos usando uma linha de nos 40 anos e nos 6000 mil reais...

```python
plt.figure(figsize=(10,6))
axes = plt.scatter(tbl["Idade"], tbl["Salário"], color="r")
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.axhline(y = 6000, color ="green", linestyle ="--")
plt.vlines(40, 0, 10000, linestyles ="--", colors ="g")
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Agora imagine que você quer trocar a cor de alguns pontos do nosso gráfico... para isso criamos um "novo layer" 

```python
plt.figure(figsize=(10,6))
axes = plt.scatter(tbl["Idade"], tbl["Salário"], color="r")
axes = plt.scatter(tbl["Idade"][0:10], tbl["Salário"][0:10], color="y")
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.axhline(y = 6000, color ="green", linestyle ="--")
plt.vlines(40, 0, 10000, linestyles ="--", colors ="g")
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Para um ponto específico podemos fazer o seguinte

```python
plt.figure(figsize=(10,6))
axes = plt.scatter(tbl["Idade"], tbl["Salário"], color="r")
axes = plt.scatter(45, 5000, color="b")
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.axhline(y = 6000, color ="green", linestyle ="--")
plt.vlines(40, 0, 10000, linestyles ="--", colors ="g")
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

Caso eu quiser aumentar o tamanho dos pontos do gráfico podemos fazer o seguinte:

```python
size = [150 for i in range(0,len(tbl["Idade"]))]
plt.figure(figsize=(10,6))
axes = plt.scatter(tbl["Idade"], tbl["Salário"], s = size, color="r")
axes = plt.scatter(45, 5000, color="b", s=[150])
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.axhline(y = 6000, color ="green", linestyle ="--")
plt.vlines(40, 0, 10000, linestyles ="--", colors ="g")
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```

E se nós quisermos alterar o símbolo? poderiamos usar setas, estrelas ou qualquer outro...

```python
size = [150 for i in range(0,len(tbl["Idade"]))]
plt.figure(figsize=(10,6))
axes = plt.scatter(tbl["Idade"], tbl["Salário"], s = size, color="r", marker='*')
axes = plt.scatter(45, 5000, color="b", s=[150], marker='>')
plt.title("Salário das pessoas com base na idade", fontsize=15)
plt.axhline(y = 6000, color ="green", linestyle ="--")
plt.vlines(40, 0, 10000, linestyles ="--", colors ="g")
plt.xlabel("Idade", fontsize=15)
plt.ylabel("Salário", fontsize=15)
plt.show()
```
