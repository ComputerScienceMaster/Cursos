Nesse notebook vamos aprender como criar grficos/diagramas com um nome bastante incomum: "Diagrama de Sankey". Para isso, vamos usar a biblioteca plotly que est disponvel para utilizao gratuita [bem aqui](https://plotly.com/python/getting-started/).

Antes de mais nada vamos tirar essa dvida, quem diabos foi Sankey?

De acordo com nossa amiga Wikipedia:
> Os diagramas de Sankey tm o nome do capito irlands Matthew *Henry Phineas Riall Sankey* , que usou esse tipo de diagrama em 1898 em uma figura clssica mostrando a eficincia energtica de uma mquina a vapor.

S por curiosidade esse  o capito Sankey:


De uma forma geral esse tipo de diagrama representa fluxo de dados, ou seja, a transferncia de alguma propriedade (seja ela fsica ou no) entre determinadas etapas. O diagrama mostra de maneira visual por meio de linhas (de espessura diferente a quantidade transferida de uma etapa para outra.

Vamos a um exemplo simples, imagine que voc deseja representar o quanto de pipoca  vendido no cinema. Ento criamos o seguinte diagrama:

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["Pipocas vendidas", "Casais", "Homens" ,"Mulheres", "Grupos de amigos", "Idosos"],
      color = "yellow"
    ),
    link = dict(
      source = [0, 0, 0, 0, 0, 0], 
      target = [1, 2, 2, 3, 4, 5],
      value = [18, 4, 2, 8, 4, 2]
  ))])

fig.update_layout(title_text="Fluxo de pipocas no cinema", font_size=15)
fig.show()
```

Finalmente, podemos dizer que a nica parte que  bastante obscura  a insero de dados... Ela realmente at o momento no faz muito sentido. No entanto, para ficar mais fcil vamos tentar traduzir esses dados para algo que  mais fcil de entender, ou seja, vamos usar tabelas. 

Ento vamos usar os mesmos dados e vamos criar um dataframe desses dados:


```python
import pandas as pd

produtosVendidos = ["Pipocas", "Chocolates"]
compradores = ["Casais", "Grupos de amigos", "Idosos", "Homens", "Mulheres"]
dados = [[2,4],[4,5],[5,6],[7,6],[6,3]]

tbl = pd.DataFrame(data = dados, columns=produtosVendidos, index = compradores)
tbl = tbl.reset_index()
tbl.head()
```

Perfeito, temos uma tabela, mas para transformar essa incrvel tabela em uma sequncia que consiga ser interpretada pelo nosso plotly, precisamos pivotar essa tabela. 

```python
tbl_t = tbl.melt(id_vars=['index'])
tbl_t
```

Perfeito, temos os valores... vamos tentar inserir no grfico?

```python
import plotly.graph_objects as go



fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["Pipocas vendidas", "Chocolates vendidos", "Casais", "Homens" ,"Mulheres", "Grupos de amigos", "Idosos"],
      color = ["rgba(255,255,0,0.5)", 
               "rgba(255,0,0,0.5)", 
               "rgba(45,0,0,0.5)", 
               "rgba(255,255,0,0.5)", 
               "rgba(255,0,0,0.5)", 
               "rgba(45,24,0,0.5)", 
               "rgba(251,45,20,0.5)", 
               "rgba(45,10,110,0.5)", 
               "rgba(55,202,14,0.5)", 
               "rgba(85,22,50,0.5)", 
               "rgba(95,102,40,0.5)", 
               "rgba(105,40,10,0.5)"]
    ),
    link = dict(
      source = tbl_t["variable"].array, 
      target = tbl_t["index"].array,
      value =  tbl_t["value"].array,
      color = ["rgba(255,255,0,0.5)", 
               "rgba(255,0,0,0.5)", 
               "rgba(45,0,0,0.5)", 
               "rgba(255,255,0,0.5)", 
               "rgba(255,0,0,0.5)", 
               "rgba(45,24,0,0.5)", 
               "rgba(251,45,20,0.5)", 
               "rgba(45,10,110,0.5)", 
               "rgba(55,202,14,0.5)", 
               "rgba(85,22,50,0.5)", 
               "rgba(95,102,40,0.5)", 
               "rgba(105,40,10,0.5)"]
  ))])

fig.update_layout(title_text="Fluxo de pipocas no cinema", font_size=15)
fig.show()
```

Apesar de todos nossos dados estarem corretos, nada apareceu, isso acontece porque ns no convertemos nossos dados que so "strings" para inteiros. Assim, o nosso plotly pode entender os dados.

```python
produtos = {'Pipocas': 1,'Chocolates': 2}

tbl_t["variable"] = [produtos[item] for item in tbl_t['variable']]
tbl_t

publico = {'Casais': 3,'Grupos de amigos': 4,'Idosos': 5, 'Homens': 6, 'Mulheres': 7}
tbl_t["index"] = [publico[item] for item in tbl_t['index']]
tbl_t
```

Agora sim, se rodarmos nosso grfico tudo sai como o esperado...

Um detalhe legal aqui  que estou escolhendo um conjunto padro do plotly disponvel [aqui](https://plotly.com/python/discrete-color/). 

O nome do conjunto  "Pastel1"

```python
import plotly.graph_objects as go
import plotly.express as px

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["Pipocas vendidas", "Chocolates vendidos", "Casais", "Homens" ,"Mulheres", "Grupos de amigos", "Idosos"],
      color = px.colors.qualitative.Pastel1
    ),
    link = dict(
      source = tbl_t["variable"].array, 
      target = tbl_t["index"].array,
      value =  tbl_t["value"].array,
      color = px.colors.qualitative.Pastel1
  ))])

fig.update_layout(title_text="Fluxo de pipocas no cinema", font_size=15)
fig.show()
```
